# Feature Engineering

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):#ast.literal_eval function helps in converting in correct format
        L.append(i['name'])
    return L
        
```

```python
data['genres'] = data['genres'].apply(convert)
```

```python
data.head()
```

```python
data['keywords'] = data['keywords'].apply(convert)
```

```python
data.head()
```

```python
# Taking only first three dictoonary from cast column that is, the main cast of the movie
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):#ast.literal_eval function helps in converting in correct format
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
data['cast'] = data['cast'].apply(convert3)
```

```python
data.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):#ast.literal_eval function helps in converting in correct format
        if (i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L
```

```python
data['crew'] = data['crew'].apply(fetch_director)
```

```python
data.head()
```

```python
#data['overview'] = data['overview'].apply(lambda x:x.split())
```

```python
#data.head()
```

```python
data['genres']=data['genres'].apply(lambda x:[i.replace(" "," ") for i in x])
data['keywords']=data['keywords'].apply(lambda x:[i.replace(" "," ") for i in x])
data['cast']=data['cast'].apply(lambda x:[i.replace(" "," ") for i in x])
data['crew']=data['crew'].apply(lambda x:[i.replace(" "," ") for i in x])
```

```python
data.head()
```

```python
data['overview'] = data['overview'].apply(lambda x:x.split())
```

```python
data['tag'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
```

```python
data.head()
```

```python
data = data[['id','tag','title']]
```

```python
data.head()
```

```python
data['tag'] = data['tag'].apply(lambda x: " ".join(x))
data.head()
```

```python
data['tag'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(data['tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = data[data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(data.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
